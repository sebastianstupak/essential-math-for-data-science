## Derivatives

Example 1-17. A derivative calculator in Python

```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2)
```

Example 1-20. Using the substitution feature in SymPy

```python
from sympy import symbols, diff, sympify

x = symbols('x')

def f(x):
    return x**2

# Create a SymPy expression from the function
f_expr = sympify(f(x))

# Take the derivative symbolically
dx_f = diff(f_expr, x)

print("Symbolic derivative:", dx_f)

slope_at_2 = dx_f.subs(x, 2)
print("Slope at x=2:", slope_at_2)
```

Example 1-22. Using limits to calculate a slope

```python
from sympy import *

# "x" and step size "s"
x, s = symbols('x s')

f = x**2

# slope between two points with gap "s"
# substitute into rise-over-run formula
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)

# substitute 2 for x
slope_2 = slope_f.subs(x, 2)

# calculate slope at x = 2
# infinitely approach step size _s_ to 0
result = limit(slope_2, s, 0)

print(result)
```

Example 1-23. Using limits to calculate a derivative

Example 1-25. Calculating the derivative dz/dx with and without the chain rule, but still getting the same answer

```python
from sympy import *

x, y = symbols('x y')

# derivative for first function
# need to underscore y to prevent variable clash
_y = x**2 + 1
dy_dx = diff(_y)

# derivative for second function
z = y**3 - 2
dz_dy = diff(z)

# calculate derivative with and without
# chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = (diff(z.subs(y, _y)))

# prove chain rule by showing both are equal
print(dz_dx_chain)
print(dz_dx_no_chain)
```

## Integrals

Example 1-26. An integral approximation in Python

```python
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0

    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)

    return total_sum * delta_x

def my_function(x):
    return x**2 + 1

area = approximate_integral(a =0, b=1, n=5, f=my_function)

print(area)
```

Example 1-27. Another integral approximation in Python

```python
area = approximate_integral(a=0, b=1, n=1000, f=my_function)

print(area)
```

Example 1-28. Yet another integral approximation in Python

```python
area = approximate_integral(a=0, b=1, n=1_000_000, f=my_function)

print(area)
```

### Calculating Integrals with Limits

Example 1-30. Using limits to calculate integrals

```python
from sympy import *

x, i, n = symbols('x i n')

f = x**2 + 1
lower, upper = 0, 1

# calculate width and each rectangle height ad index "i"
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)

# iterate all "n" rectangles and sum their areas
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# calcualte the area by approaching the number
# of rectangles "n" to infinity
area = limit(n_rectangles, n, oo)

print(area)
```

5. Question

Q: Assuming no payments are made, how much would a $1,000 loan be worth at 5% interest compounded monthly after 3 years?

A: For compound interest with monthly compounding, we use the formula:

`A = P(1 + r/n)^(nt)`

Where:
- P = Principal ($1,000)
- r = Annual interest rate (5% = 0.05)
- n = Number of times compounded per year (12 for monthly)
- t = Time in years (3)

```python
principal = 1000
rate = 0.05
time = 3
n = 12  # compounds monthly

amount = principal * (1 + rate/n)**(n*time)
print(f"${amount:.2f}")
```

6. Question

Q: Assuming no payments are made, how much would a $1,000 loan be worth at 5% interest compounded continuously after 3 years?

A: For continuous compounding, we use a different formula:

`A = P * e^(rt)`

Where:
- P = Principal ($1,000)
- r = Annual interest rate (5% = 0.05)
- t = Time in years (3)
- e = Euler's number

```python
import math

principal = 1000
rate = 0.05
time = 3

amount = principal * math.exp(rate * time)
print(f"${amount:.2f}")
```

7. Question

Q: For the function f(x) = 3x^2 + 1 what is the slope at x = 3?

A: 

```python
from sympy import *

# Calculate the derivative
f_prime = diff(f, x)
print(f"Derivative f'(x) = {f_prime}")

# Calculate slope at x = 3
slope = f_prime.subs(x, 3)
print(f"Slope at x = 3: {slope}")

# Calculate point on curve at x = 3
point_y = f.subs(x, 3)

# Create tangent line equation: y = mx + b
# y - y1 = m(x - x1)
# y = mx - mx1 + y1
tangent = slope*(x - 3) + point_y

# Plot both functions
plot(f, tangent, (x, -1, 5))
```

8. Question

Q: For the function f(x) = 3x^2 + 1 what is the area under the curve for x between 0 and 2?

A:

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
f = 3*x**2 + 1
area = integrate(f, (x, 0, 2))
print(f"Area under the curve from x = 0 to x = 2: {area}")

x_vals = np.linspace(0, 2, 100)
y_vals = 3*x_vals**2 + 1

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', label='f(x) = 3x + 1')
plt.fill_between(x_vals, y_vals, alpha=0.3)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area under f(x) = 3x + 1 from x=0 to x=2')
plt.legend()
plt.show()
```
